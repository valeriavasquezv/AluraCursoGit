# **Proyecto** **Telecom** **X**

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/valeriavasquezv/valeriavasquez-portfolio/refs/heads/main/Proyecto_TelecomX/TelecomX_Data.json"
```

## EXTRACCION DE LOS DATOS

```python
df_telecom = pd.read_json(url)
df_telecom
```

```python
df_telecom.info()
```

```python
df_telecom.dtypes
```

##TRANSFORMACIÓN DE LOS DATOS


```python
churn_df = df_telecom['Churn']
pd.unique(churn_df)

#columna churn contiene valores en blanco, revisar en siguiente etapa
```

```python
customer_df = pd.json_normalize(df_telecom["customer"])

pd.unique(customer_df["gender"])
pd.unique(customer_df["Partner"])
pd.unique(customer_df["Dependents"])
pd.unique(customer_df["SeniorCitizen"])
pd.unique(customer_df['tenure'])
```

```python
phone_df = pd.json_normalize(df_telecom['phone'])

pd.unique(phone_df['PhoneService'])
pd.unique(phone_df['MultipleLines'])
```

```python
internet_df = pd.json_normalize(df_telecom['internet'])

pd.unique(internet_df['InternetService'])
pd.unique(internet_df['OnlineSecurity'])
pd.unique(internet_df['OnlineBackup'])
pd.unique(internet_df['DeviceProtection'])
pd.unique(internet_df['TechSupport'])
pd.unique(internet_df['StreamingTV'])
pd.unique(internet_df['StreamingMovies'])

```

```python
account_df = pd.json_normalize(df_telecom['account'])

pd.unique(account_df['Contract'])
pd.unique(account_df['PaperlessBilling'])
pd.unique(account_df['PaymentMethod'])
pd.unique(account_df['Charges.Monthly'])
pd.unique(account_df['Charges.Total'])
```
